### ðŸ”— **Daten zusammenfÃ¼hren (Merge & Join in Pandas)**

**Ziel:**

Verstehen, wie man mehrere Tabellen anhand gemeinsamer Spalten zusammenfÃ¼hrt (wie in Excel VLOOKUP oder SQL JOIN).

**Wichtige Methode:**

pd.merge(df1, df2, on="gemeinsame_Spalte", **how**="inner")

`how` kann sein:

- `"inner"` â€“ nur gemeinsame Werte (Standard)
- `"left"` â€“ alle aus der linken Tabelle
- `"right"` â€“ alle aus der rechten Tabelle
- `"outer"` â€“ alle Werte, auch wenn sie nicht Ã¼bereinstimmen

```python
import pandas as pd

# Erste Tabelle: Personen
personen = pd.DataFrame({
    "Name": ["Anna", "Ben", "Clara", "David"],
    "Abteilungs_ID": [1, 2, 2, 3]
})

# Zweite Tabelle: Abteilungen
abteilungen = pd.DataFrame({
    "Abteilungs_ID": [1, 2, 3],
    "Abteilung": ["HR", "Data", "Marketing"]
})

# Tabellen zusammenfÃ¼hren
merged = pd.merge(personen, abteilungen, on="Abteilungs_ID", how="inner")

print("âž¡ï¸ Ergebnis:")
print(merged)
```
