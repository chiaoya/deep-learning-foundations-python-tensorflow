## üß© For-Schleife

```python
# Zahlen von 0 bis 5 ausgeben
for i in range(6):
    print("Zahl:", i)
```

```python
# Zahlen von 1 bis 5 ausgeben
for i in range(1, 6):
    print("Zahl:", i)
```

**Zahlen von 0 bis einschlie√ülich 10 mit Schrittweite 2 ausgeben**

range(start, stop, step) erzeugt eine Zahlenfolge

- start = 0 ‚Üí Startwert

- stop = 10 ‚Üí 10 ist nicht enthalten

- step = 2 ‚Üí Schrittweite 2

```python
# Zahlen von 0 bis einschlie√ülich 10 mit Schrittweite 2 ausgeben
for i in range(0, 10, 2) :
    print("Zahl:", i)
```

```python
farben = ["Rot", "Gr√ºn", "Blau"]
for farbe in farben:
    print("Ich mag die Farbe", farbe)
```

## üß© While-Schleife

```python
# Countdown von 5 bis 1
x = 5
while x > 0:
    print("Countdown:", x)
    x -= 1 # x = x-1
print("Start!")
```

### üîó **Daten zusammenf√ºhren (Merge & Join in Pandas)**

**Ziel:**

Verstehen, wie man mehrere Tabellen anhand gemeinsamer Spalten zusammenf√ºhrt (wie in Excel VLOOKUP oder SQL JOIN).

**Wichtige Methode:**

pd.merge(df1, df2, on="gemeinsame_Spalte", **how**="inner")

`how` kann sein:

- `"inner"` ‚Äì nur gemeinsame Werte (Standard)
- `"left"` ‚Äì alle aus der linken Tabelle
- `"right"` ‚Äì alle aus der rechten Tabelle
- `"outer"` ‚Äì alle Werte, auch wenn sie nicht √ºbereinstimmen

```python
import pandas as pd

# Erste Tabelle: Personen
personen = pd.DataFrame({
    "Name": ["Anna", "Ben", "Clara", "David"],
    "Abteilungs_ID": [1, 2, 2, 3]
})

# Zweite Tabelle: Abteilungen
abteilungen = pd.DataFrame({
    "Abteilungs_ID": [1, 2, 3],
    "Abteilung": ["HR", "Data", "Marketing"]
})

# Tabellen zusammenf√ºhren
merged = pd.merge(personen, abteilungen, on="Abteilungs_ID", how="inner")

print("‚û°Ô∏è Ergebnis:")
print(merged)
```
